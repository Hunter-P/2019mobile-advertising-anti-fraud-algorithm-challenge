### 通过ip得到地理，运营商信息

```python
data.columns
```

```python
from ip2Region import Ip2Region

def ip2city(ip):
    db_file = "ip2region.db"
    searcher = Ip2Region(db_file)
    try:
        data = searcher.memorySearch(ip)
        return data["region"].decode('utf-8').split("|")
    except:
        return [0, 0, 0, 0]

ip2city("106.121.13.8")
```

```python
ip_info = []
for ip in data.ip:
    ip_info.append(ip2city(ip))
ip_info = pd.DataFrame(ip_info, columns=['country', '0', 'province', 'city', 'carrier'])
ip_info
```

```python
ip_info.to_csv('8000000_ip_info.csv', index=False)
# data['ip_country'] = ip_info.country
# data['ip_province'] = ip_info.province
# ind = (ip_info.city=='0')
# ip_info['city'][ind]=data['city'][ind]
# data['city'] = ip_info['city']
```

```python
reqrealip_info = []
for ip in data.reqrealip:
    reqrealip_info.append(ip2city(ip))
reqrealip_info = pd.DataFrame(reqrealip_info, columns=['country', 'unknown', 'province', 'city', 'carrier'])
reqrealip_info.to_csv('reqrealip_info.csv', index=False)
data['reqrealip_province'] = reqrealip_info.province
data['reqrealip_city'] = reqrealip_info.city
data['reqrealip_carrier'] = reqrealip_info.carrier
```
